I will first rename the 'birth year' column to 'birth_year' to avoid syntax issues in the statsmodels formula. Then, I will define the `run_regression` function as requested, which performs WLS regression with clustered standard errors.



```python
# Rename 'birth year' to 'birth_year' for easier use in formulas
df_merged_final = df_merged_final.rename(columns={'birth year': 'birth_year'})

def run_regression(outcome_var, eligibility_var, data):
    """
    Runs a WLS regression of outcome_var on eligibility_var with controls and fixed effects.

    Args:
    outcome_var (str): Name of the outcome variable.
    eligibility_var (str): Name of the eligibility variable.
    data (pd.DataFrame): The dataframe containing the data.

    Returns:
    tuple: (coefficient, standard error, p-value) for the eligibility variable.
    """

    # Construct the formula
    # Note: We use C() to treat variables as categorical fixed effects
    formula = f"{outcome_var} ~ {eligibility_var} + C(STATE_BPL) + C(birth_year) + C(AGE) + C(SEX) + C(RACE) + C(HISPAN)"

    # Fit the WLS model
    # We use 'PERWT' as weights (Person Weights)
    model = smf.wls(formula=formula, data=data, weights=data['PERWT'])

    # Fit with clustered standard errors on 'STATE_BPL'
    results = model.fit(cov_type='cluster', cov_kwds={'groups': data['STATE_BPL']})

    # Extract results for the eligibility variable
    coef = results.params[eligibility_var]
    std_err = results.bse[eligibility_var]
    p_value = results.pvalues[eligibility_var]

    return coef, std_err, p_value

print("Function 'run_regression' defined.")
print("Column 'birth year' renamed to 'birth_year'.")
```

```python
def run_economic_trend_regression(outcome_var, eligibility_var, data):
    """
    Runs a WLS regression of outcome_var on eligibility_var with economic controls,
    linear time trend, and fixed effects.

    Args:
    outcome_var (str): Name of the outcome variable.
    eligibility_var (str): Name of the eligibility variable.
    data (pd.DataFrame): The dataframe containing the data.

    Returns:
    statsmodels.regression.linear_model.RegressionResultsWrapper: The fitted regression results.
    """

    # Construct the formula including economic controls and YEAR as a trend
    # Formula: Outcome ~ Eligibility + Economic_Controls + Linear_Year + Fixed_Effects
    formula = f"{outcome_var} ~ {eligibility_var} + Childhood_Avg_Unemployment + log_childhood_income + YEAR + C(STATE_BPL) + C(AGE) + C(SEX) + C(RACE) + C(HISPAN)"

    # Fit the WLS model
    # We use 'PERWT' as weights (Person Weights)
    model = smf.wls(formula=formula, data=data, weights=data['PERWT'])

    # Fit with clustered standard errors on 'STATE_BPL'
    results = model.fit(cov_type='cluster', cov_kwds={'groups': data['STATE_BPL']})

    return results

print("Function 'run_economic_trend_regression' defined.")
```

```python
def run_wls_regression(outcome_var, eligibility_var, data):
    """
    Runs a WLS regression of outcome_var on eligibility_var with economic controls,
    linear time trend, and fixed effects, returning specific statistics.

    Args:
    outcome_var (str): Name of the outcome variable.
    eligibility_var (str): Name of the eligibility variable.
    data (pd.DataFrame): The dataframe containing the data.

    Returns:
    tuple: (elig_coef, elig_se, elig_pvalue, rsquared, nobs) for the eligibility variable and model.
    """

    # Construct the formula including economic controls and YEAR as a trend
    # Formula: Outcome ~ Eligibility + Economic_Controls + Linear_Year + Fixed_Effects
    formula = f"{outcome_var} ~ {eligibility_var} + Childhood_Avg_Unemployment + log_childhood_income + YEAR + C(STATE_BPL) + C(AGE) + C(SEX) + C(RACE) + C(HISPAN)"

    # Fit the WLS model
    # We use 'PERWT' as weights (Person Weights)
    model = smf.wls(formula=formula, data=data, weights=data['PERWT'])

    # Fit with clustered standard errors on 'STATE_BPL'
    results = model.fit(cov_type='cluster', cov_kwds={'groups': data['STATE_BPL']})

    # Extract results for the eligibility variable
    elig_coef = results.params[eligibility_var]
    elig_se = results.bse[eligibility_var]
    elig_pvalue = results.pvalues[eligibility_var]

    # Extract R-squared and number of observations
    rsquared = results.rsquared
    nobs = results.nobs

    return elig_coef, elig_se, elig_pvalue, rsquared, nobs

print("Function 'run_wls_regression' defined.")
```

```python
def run_wls_regression(outcome_var, eligibility_var, data):
    """
    Runs a WLS regression of outcome_var on eligibility_var with economic controls,
    linear time trend, and fixed effects, returning specific statistics.

    Args:
    outcome_var (str): Name of the outcome variable.
    eligibility_var (str): Name of the eligibility variable.
    data (pd.DataFrame): The dataframe containing the data.

    Returns:
    tuple: (elig_coef, elig_se, elig_pvalue, rsquared, nobs) for the eligibility variable and model.
    """

    # Define all variables involved in the regression (outcome, eligibility, controls, weights, clustering groups)
    all_vars_for_model = [
        outcome_var, eligibility_var, 'Childhood_Avg_Unemployment',
        'log_childhood_income', 'YEAR', 'STATE_BPL', 'AGE', 'SEX',
        'RACE', 'HISPAN', 'PERWT' # PERWT for weights, STATE_BPL for clustering
    ]

    # Create a temporary dataframe with only the necessary columns and drop rows with any NaNs in these columns
    # This ensures alignment between data, weights, and clustering groups
    temp_data = data[all_vars_for_model].dropna().copy()

    # Construct the formula including economic controls and YEAR as a trend
    # Formula: Outcome ~ Eligibility + Economic_Controls + Linear_Year + Fixed_Effects
    formula = f"{outcome_var} ~ {eligibility_var} + Childhood_Avg_Unemployment + log_childhood_income + YEAR + C(STATE_BPL) + C(AGE) + C(SEX) + C(RACE) + C(HISPAN)"

    # Fit the WLS model
    # We use 'PERWT' as weights (Person Weights)
    model = smf.wls(formula=formula, data=temp_data, weights=temp_data['PERWT'])

    # Fit with clustered standard errors on 'STATE_BPL'
    # The groups should come from the same temp_data to ensure alignment
    results = model.fit(cov_type='cluster', cov_kwds={'groups': temp_data['STATE_BPL']})

    # Extract results for the eligibility variable
    elig_coef = results.params[eligibility_var]
    elig_se = results.bse[eligibility_var]
    elig_pvalue = results.pvalues[eligibility_var]

    # Extract R-squared and number of observations
    rsquared = results.rsquared
    nobs = results.nobs

    return elig_coef, elig_se, elig_pvalue, rsquared, nobs

print("Function 'run_wls_regression' defined.")
```
