````python
import numpy as np

# select relevant columns
age_cols = [
    'total_15', 'total_16', 'total_17', 'total_18', 'total_19',
    'total_20', 'total_21', 'total_22', 'total_23', 'total_24',
    'total_25_29', 'total_30_34', 'total_35_39', 'total_40_44',
    'total_45_49', 'total_50_54', 'total_55_59', 'total_60_64',
    'total_over_64'
]

id_vars = ['fips_state_code', 'year']

# Melt the dataframe
melted_df = arrests_df.melt(id_vars=id_vars, value_vars=age_cols, var_name='age_group', value_name='crime_count')

# Clean up the age column
# We need to extract a numeric age representing the cohort.
# For single ages (e.g., total_15), it's straightforward.
# For ranges (e.g., total_25_29), we will take the lower bound as a proxy for the cohort start,
# or essentially treat '25' as the age to subtract from year to get a rough birth year cohort.
# Given the task asks for `brthyr = year - age`, using the integer start of the bin is the most logical interpretation for cohort matching.

def extract_age(s):
    # Remove 'total_'
    s = s.replace('total_', '')
    # If it's a range like '25_29', take the first part '25'
    if '_' in s:
        s = s.split('_')[0]
    # Handle 'over_64' -> 65 (arbitrary choice, but usually treated as such or excluded)
    if s == 'over': # resulting from split of over_64 -> ['over', '64']
         return 65
    return int(s)

melted_df['age'] = melted_df['age_group'].apply(extract_age)

# Rename fips_state_code to fips
melted_df = melted_df.rename(columns={'fips_state_code': 'fips'})

# Aggregate sum of crime counts
aggregated_arrests = melted_df.groupby(['fips', 'year', 'age'])['crime_count'].sum().reset_index()

# Calculate birth year
aggregated_arrests['brthyr'] = aggregated_arrests['year'] - aggregated_arrests['age']

print(aggregated_arrests.head())
print(aggregated_arrests.describe())
````
