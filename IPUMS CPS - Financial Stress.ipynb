Inspect the already loaded dataframe structure (shape, types, years).

```python
print("Shape of the dataframe:")
print(df_acs.shape)
print("\nColumn names:")
print(df_acs.columns)
print("\nData types:")
print(df_acs.dtypes)

print("\nYear range:")
print(f"Min Year: {df_acs['YEAR'].min()}")
print(f"Max Year: {df_acs['YEAR'].max()}")
```

```python
# --- Step 1: Detailed Exploration and Column Check ---

# List of columns to check
check_cols = [
    'YEAR', 'SERIAL', 'PERNUM', 'ASECWT', 'AGE', 'SEX', 'RACE', 'HISPAN', 'STATEFIP',
    'RELATE', 'MOMLOC', 'POPLOC', 'FAMSIZE', 'NCHILD', 'HHINCOME', 'OFFPOV', 'FOODSTMP',
    'STAMPVAL', 'INCWELFR', 'HIMCAIDLY', 'HCOVPRIV', 'PHINSUR', 'COVERGH',
    'EMPSTAT', 'WKSWORK1', 'UHRSWORKLY', 'INCWAGE', 'CLASSWKR'
]
# Note: Adjusted some variable names to match what I saw in df_acs.columns earlier
# (e.g., INCWELFR vs INCWELF, UHRSWORKLY vs UHRSWORK)

missing_cols = [c for c in check_cols if c not in df_acs.columns]
if missing_cols:
    print(f"WARNING: The following columns are MISSING: {missing_cols}")
else:
    print("SUCCESS: All key columns are present.")

print("\n--- Descriptive Stats for AGE ---")
print(df_acs['AGE'].describe())

print("\n--- Value Counts for RELATE (Top 20) ---")
print(df_acs['RELATE'].value_counts().sort_index().head(20))

print("\n--- Value Counts for MOMLOC (Top 10) ---")
print(df_acs['MOMLOC'].value_counts().sort_index().head(10))

print("\n--- Value Counts for POPLOC (Top 10) ---")
print(df_acs['POPLOC'].value_counts().sort_index().head(10))
```

```python
# --- Step 2: Create birth_year and Initial Filter ---

# 1. Create birth_year
df_acs['birth_year'] = df_acs['YEAR'] - df_acs['AGE']

# 2. Print distribution
print("\n--- Birth Year Distribution (Sample) ---")
print(df_acs['birth_year'].value_counts().sort_index().loc[1975:1990])

# 3. Filter for 1981-1984 cohorts
target_cohorts = [1981, 1982, 1983, 1984]
df_cohorts = df_acs[df_acs['birth_year'].isin(target_cohorts)].copy()

print(f"\nOriginal Shape: {df_acs.shape}")
print(f"Shape after Cohort Filter (1981-1984): {df_cohorts.shape}")
```

```python
# --- Step 3: Identify Children using Multiple Criteria ---

# Criterion A: Age <= 18
df_cohorts['is_child_age'] = (df_cohorts['AGE'] <= 18).astype(int)

# Criterion B: Relationship (301=Child, 303=Stepchild)
# Adjust codes if RELATE inspection shows different standard (e.g. 301 vs 3)
df_cohorts['is_child_relate'] = df_cohorts['RELATE'].isin([301, 303, 1114]).astype(int) # Added 1114 (Grandchild) just in case, but primary is 301/303

# Criterion C: Parent Pointer
df_cohorts['is_child_parent_ptr'] = ((df_cohorts['MOMLOC'] > 0) | (df_cohorts['POPLOC'] > 0)).astype(int)

# Cross-tabs
print("\n--- Cross-tab: Age <= 18 vs. Relate --- ভা")
print(pd.crosstab(df_cohorts['is_child_age'], df_cohorts['is_child_relate']))

print("\n--- Cross-tab: Age <= 18 vs. Parent Pointer ---")
print(pd.crosstab(df_cohorts['is_child_age'], df_cohorts['is_child_parent_ptr']))

# Final Child Indicator: Age <= 18 AND Parent Present
df_cohorts['is_child'] = ((df_cohorts['is_child_age'] == 1) & (df_cohorts['is_child_parent_ptr'] == 1)).astype(int)

print("\n--- Final Child Indicator Counts ---")
print(df_cohorts['is_child'].value_counts())

print("\n--- Age Distribution of Identified Children ---")
print(df_cohorts[df_cohorts['is_child'] == 1]['AGE'].value_counts().sort_index())

# Filter to keep only identified children
df_children = df_cohorts[df_cohorts['is_child'] == 1].copy()
print(f"\nFinal Shape (df_children): {df_children.shape}")
```

```python
# --- Step 4: Validate the Child Sample ---

# 1. Cohort x Year Coverage
print("\n--- Cohort Coverage (Birth Year x Survey Year) ---")
# We expect a diagonal pattern where age increases with year for each cohort
crosstab_coverage = pd.crosstab(df_children['birth_year'], df_children['YEAR'])
display(crosstab_coverage)

# 2. Age at Survey Stats
df_children['age_at_survey'] = df_children['AGE']
print("\n--- Age at Survey Stats ---")
print(df_children['age_at_survey'].describe())

# 3. Obs per Birth Year
print("\n--- Observations per Birth Year ---")
print(df_children['birth_year'].value_counts().sort_index())

# 4. State Coverage
print("\n--- Observations per State (Top 5 & Bottom 5) ---")
state_counts = df_children['STATEFIP'].value_counts()
print("Top 5 States by Count:")
print(state_counts.head(5))
print("\nBottom 5 States by Count:")
print(state_counts.tail(5))

# Check for states with very low observations
low_obs_states = state_counts[state_counts < 50]
if not low_obs_states.empty:
    print(f"\nWARNING: The following states have < 50 observations: {low_obs_states.index.tolist()}")
else:
    print("\nSUCCESS: All states have >= 50 observations.")

# 5. Observations per Year
print("\n--- Observations per Survey Year ---")
print(df_children['YEAR'].value_counts().sort_index())

# 6. Filter Non-State Geographies (STATEFIP > 56)
# (Codes > 56 are usually territories or errors in this context)
n_before_state = len(df_children)
df_children = df_children[df_children['STATEFIP'] <= 56].copy()
n_dropped_state = n_before_state - len(df_children)

print(f"\nDropped {n_dropped_state} observations with STATEFIP > 56.")
print(f"Final Shape after Step 4: {df_children.shape}")
```

```python
# --- Step 5: Link Parent Characteristics ---

# define columns to link
# Note: INCWAGE was missing in initial check. Using UHRSWORKLY instead of UHRSWORK.
parent_cols = ['YEAR', 'SERIAL', 'PERNUM', 'EMPSTAT', 'WKSWORK1', 'UHRSWORKLY', 'CLASSWKR']

# Create lookup table from ORIGINAL df_acs (before filtering)
# We need to ensure we don't have duplicates on keys, though CPS structure usually guarantees this per year/serial/pernum
parent_lookup = df_acs[parent_cols].copy()

# 1. Merge Mother's Characteristics
# Join on YEAR, SERIAL, and MOMLOC (Child) == PERNUM (Parent)
df_children = df_children.merge(
    parent_lookup,
    left_on=['YEAR', 'SERIAL', 'MOMLOC'],
    right_on=['YEAR', 'SERIAL', 'PERNUM'],
    how='left',
    suffixes=('', '_mom')
)

# Rename the newly added columns
rename_mom = {
    'EMPSTAT': 'mom_EMPSTAT',
    'WKSWORK1': 'mom_WKSWORK1',
    'UHRSWORKLY': 'mom_UHRSWORKLY',
    'CLASSWKR': 'mom_CLASSWKR',
    'PERNUM': 'mom_PERNUM_check' # Just for verification, can drop later
}
df_children.rename(columns=rename_mom, inplace=True)

# 2. Merge Father's Characteristics
# Join on YEAR, SERIAL, and POPLOC (Child) == PERNUM (Parent)
df_children = df_children.merge(
    parent_lookup,
    left_on=['YEAR', 'SERIAL', 'POPLOC'],
    right_on=['YEAR', 'SERIAL', 'PERNUM'],
    how='left',
    suffixes=('', '_dad')
)

# Rename the newly added columns (merges usually add suffixes to collisions, but here we control it)
# Since 'EMPSTAT' etc were renamed in the previous step, the new ones coming in might have no suffix or '_dad' depending on overlap.
# Let's handle the specific naming resulting from the merge.

# The merge above likely created columns like 'EMPSTAT', 'WKSWORK1' again because the left side keys were consumed.
# Let's manually rename the columns that were just added.
rename_dad = {
    'EMPSTAT': 'dad_EMPSTAT',
    'WKSWORK1': 'dad_WKSWORK1',
    'UHRSWORKLY': 'dad_UHRSWORKLY',
    'CLASSWKR': 'dad_CLASSWKR',
    'PERNUM': 'dad_PERNUM_check'
}
df_children.rename(columns=rename_dad, inplace=True)

# 3. Check Merge Success
print("\n--- Parent Merge Success Rates ---")
n_total = len(df_children)
n_mom = df_children['mom_EMPSTAT'].notna().sum()
n_dad = df_children['dad_EMPSTAT'].notna().sum()
n_both = ((df_children['mom_EMPSTAT'].notna()) & (df_children['dad_EMPSTAT'].notna())).sum()
n_neither = ((df_children['mom_EMPSTAT'].isna()) & (df_children['dad_EMPSTAT'].isna())).sum()

print(f"Total Children: {n_total}")
print(f"Linked to Mother: {n_mom} ({n_mom/n_total:.1%})")
print(f"Linked to Father: {n_dad} ({n_dad/n_total:.1%})")
print(f"Linked to BOTH:   {n_both} ({n_both/n_total:.1%})")
print(f"Linked to NEITHER:{n_neither} ({n_neither/n_total:.1%})")

# Check for rows where we expected a link (MOMLOC>0) but didn't get one
missing_mom_link = df_children[(df_children['MOMLOC'] > 0) & (df_children['mom_EMPSTAT'].isna())]
if len(missing_mom_link) > 0:
    print(f"\nWARNING: {len(missing_mom_link)} rows have MOMLOC > 0 but failed to link mother data.")
    print("Example failing rows:")
    display(missing_mom_link[['YEAR', 'SERIAL', 'PERNUM', 'MOMLOC', 'POPLOC']].head())
else:
    print("\nSUCCESS: All MOMLOC pointers successfully linked.")
```

```python
# --- Step 6: Household Structure Indicators ---

# 1. Construct Indicators
# single_parent: 1 if exactly one parent is present (MOMLOC>0 xor POPLOC>0)
# We already know at least one is present from Step 3.
df_children['single_parent'] = (
    ((df_children['MOMLOC'] > 0) & (df_children['POPLOC'] == 0)) |
    ((df_children['MOMLOC'] == 0) & (df_children['POPLOC'] > 0))
).astype(int)

df_children['num_parents'] = (df_children['MOMLOC'] > 0).astype(int) + (df_children['POPLOC'] > 0).astype(int)

# 2. Structure Statistics
print("\n--- Household Structure Stats ---")
print(f"Single Parent Rate: {df_children['single_parent'].mean():.2%}")
print("\nNumber of Parents in Household (Value Counts):")
print(df_children['num_parents'].value_counts().sort_index())

print("\n--- Single Parent Rate by Birth Year ---")
print(df_children.groupby('birth_year')['single_parent'].mean())

print("\n--- Single Parent Rate by Survey Year ---")
# Printing full series to see trends over time
print(df_children.groupby('YEAR')['single_parent'].mean())
```

```python
# --- Step 7: Final Summary and Save ---

print("\n============================================")
print("       FINAL CHILD SAMPLE SUMMARY")
print("============================================")
print(f"Total Observations: {len(df_children)}")
print(f"Unique Children (based on YEAR+SERIAL+PERNUM): {len(df_children)}")
print(f"Unique Households (based on YEAR+SERIAL): {df_children[['YEAR', 'SERIAL']].drop_duplicates().shape[0]}")

print("\n--- Demographics ---")
print("Observations per Birth Year:")
print(df_children['birth_year'].value_counts().sort_index())

print("\nAge Distribution:")
print(df_children['AGE'].value_counts().sort_index())

print("\n--- Geographic Coverage ---")
print(f"Number of Unique States: {df_children['STATEFIP'].nunique()}")
print("Top 10 States by Observations:")
print(df_children['STATEFIP'].value_counts().head(10))
print("Bottom 10 States by Observations:")
print(df_children['STATEFIP'].value_counts().tail(10))

print(f"\nNumber of Unique Survey Years: {df_children['YEAR'].nunique()}")

print("\n--- Financial Conditions (Means) ---")
# Check if columns exist before printing
fin_cols = ['HHINCOME', 'OFFPOV', 'FOODSTMP', 'single_parent']
existing_fin_cols = [c for c in fin_cols if c in df_children.columns]
print(df_children[existing_fin_cols].mean())

print("\n--- Parent Employment (Means of Linked Parents) ---")
# mom_EMPSTAT: 10-12=Employed, 20-22=Unemployed, 30+=Not in Labor Force (checking if < 20 for 'Employed')
if 'mom_EMPSTAT' in df_children.columns:
    print(f"Mom Employment Rate (EMPSTAT < 20): {(df_children['mom_EMPSTAT'] < 20).mean():.2%}")
if 'dad_EMPSTAT' in df_children.columns:
    print(f"Dad Employment Rate (EMPSTAT < 20): {(df_children['dad_EMPSTAT'] < 20).mean():.2%}")

# Save
output_filename = 'df_children.csv'
df_children.to_csv(output_filename, index=False)
print(f"\nSUCCESS: Final dataframe saved to {output_filename}")
print(f"Final Shape: {df_children.shape}")
```

```python
import pandas as pd
import numpy as np

# --- Step 1: Load and process annual eligibility data from .dta file ---
file_path_annual_eligibility = '/content/drive/MyDrive/Colab Notebooks/Thesis/Data/BKL Calculator/Datafortheweb/BKL.Medicaid.Calculator/medicaidcalc_cpsinst_bkl.dta'
df_annual_eligibility = pd.read_stata(file_path_annual_eligibility)

print("Loaded df_annual_eligibility from .dta file.")

# --- Step 2: Calculate cumulative eligibility from annual data ---
# Select the columns fips, brthyr, age, and sim_eligfpl
columns_to_select = ['fips', 'brthyr', 'age', 'sim_eligfpl']
df_cumulative_annual_eligibility = df_annual_eligibility[columns_to_select].copy()

# Drop any rows with NaN values in selected columns
df_cumulative_annual_eligibility.dropna(subset=columns_to_select, inplace=True)

# Convert key identifier columns to integer data types
df_cumulative_annual_eligibility['fips'] = df_cumulative_annual_eligibility['fips'].astype(int)
df_cumulative_annual_eligibility['brthyr'] = df_cumulative_annual_eligibility['brthyr'].astype(int)
df_cumulative_annual_eligibility['age'] = df_cumulative_annual_eligibility['age'].astype(int)

# Sort the DataFrame by fips, brthyr, and age
df_cumulative_annual_eligibility.sort_values(by=['fips', 'brthyr', 'age'], inplace=True)

# Remove duplicate rows based on the grouping keys
df_cumulative_annual_eligibility.drop_duplicates(subset=['fips', 'brthyr', 'age'], inplace=True)

# Group by fips and brthyr, then calculate the cumulative sum of sim_eligfpl
df_cumulative_annual_eligibility['cumulative_eligibility_to_date'] = \
    df_cumulative_annual_eligibility.groupby(['fips', 'brthyr'])['sim_eligfpl'].cumsum()

# Rename columns to match df_merged
df_cumulative_annual_eligibility.rename(columns={
    'fips': 'STATEFIP',
    'brthyr': 'birth_year',
    'age': 'age_child'
}, inplace=True)

print("Calculated df_cumulative_annual_eligibility.")
print(df_cumulative_annual_eligibility.head())

# --- Step 3: Merge new cumulative eligibility into main DataFrame ---
# Store the number of rows before merge for verification
pre_merge_rows = len(df_merged)

# Drop the existing 'cumulative_eligibility_at_age' column from df_merged if it exists
if 'cumulative_eligibility_at_age' in df_merged.columns:
    df_merged.drop(columns=['cumulative_eligibility_at_age'], inplace=True)
    print("Dropped existing 'cumulative_eligibility_at_age' from df_merged.")

# Perform a left merge of df_merged with the new cumulative annual eligibility
# Select only the necessary columns from df_cumulative_annual_eligibility for merging
subset_df_cumulative_annual_eligibility = df_cumulative_annual_eligibility[[
    'STATEFIP', 'birth_year', 'age_child', 'cumulative_eligibility_to_date'
]]
df_merged = pd.merge(
    df_merged,
    subset_df_cumulative_annual_eligibility,
    on=['STATEFIP', 'birth_year', 'age_child'],
    how='left'
)

# Rename the newly merged column 'cumulative_eligibility_to_date' in df_merged
df_merged.rename(columns={'cumulative_eligibility_to_date': 'cumulative_eligibility_at_age'}, inplace=True)

# Store the number of rows in df_merged after the merge
post_merge_rows = len(df_merged)

# Compare pre_merge_rows and post_merge_rows to verify no row inflation
print(f"\nRows before merge: {pre_merge_rows:,}")
print(f"Rows after merge: {post_merge_rows:,}")
if pre_merge_rows == post_merge_rows:
    print("SUCCESS: No row inflation or deflation occurred during the merge.")
else:
    print("WARNING: Row count changed! Investigate possible issues.")

# Calculate and print the count of non-null values for the 'cumulative_eligibility_at_age' column
n_matched_elig_new = df_merged['cumulative_eligibility_at_age'].notna().sum()
n_total_elig_new = len(df_merged)
print(f"\nNumber of non-null 'cumulative_eligibility_at_age' values: {n_matched_elig_new:,} out of {n_total_elig_new:,} ({n_matched_elig_new/n_total_elig_new:.1%})\n")

print("Updated 'cumulative_eligibility_at_age' in df_merged with non-averaged annual data.")
```

```python
fixed_effect_vars.remove('birth_year')
control_vars.append('birth_year')

print("Updated outcome variables:", outcome_vars)
print("Updated eligibility variables:", eligibility_vars)
print("Updated control variables:", control_vars)
print("Updated fixed effect variables:", fixed_effect_vars)
```

**Reasoning**:
Now that the variable lists have been modified to treat 'birth_year' as a linear control variable, the next step is to re-execute the regression analysis using these updated lists. This involves recreating the `df_reg` and `df_reg_clean` DataFrames to reflect the new variable categories and then running the WLS regressions.



```python
import statsmodels.formula.api as smf
import statsmodels.regression.linear_model as sm

# 2. Re-create a list regression_cols by combining all unique column names with updated lists
regression_cols = list(set(
    outcome_vars +
    eligibility_vars +
    control_vars +
    fixed_effect_vars +
    ['ASECWT'] # Add ASECWT explicitly
))

# 3. Re-create df_reg by selecting only the regression_cols from df_merged.
df_reg = df_merged[regression_cols].copy()

# 4. Re-create df_reg_clean by dropping rows from df_reg that have NaN values
#    in any of the control_vars or eligibility_vars. Also, ensure ASECWT is not NaN.
columns_to_check_for_nan = control_vars + eligibility_vars + ['ASECWT']
df_reg_clean = df_reg.dropna(subset=columns_to_check_for_nan).copy()

# 5. Convert the columns specified in fixed_effect_vars in df_reg_clean to the 'category' data type.
for col in fixed_effect_vars:
    df_reg_clean[col] = df_reg_clean[col].astype('category')

# 6. Print the total number of rows in df_reg_clean.
print(f"Total number of rows in df_reg_clean (re-calculated): {len(df_reg_clean)}")

# 7. Re-initialize results_list and rerun regressions
results_list = []

# 8. Loop through each outcome in outcome_vars and then through each elig in eligibility_vars:
for outcome in outcome_vars:
    for elig in eligibility_vars:
        # Inside the loops, create a temporary DataFrame data by dropping rows from df_reg_clean
        # where the current outcome variable is NaN.
        data = df_reg_clean.dropna(subset=[outcome]).copy()

        # Construct the regression formula dynamically as a string.
        # Fixed effects are added using 'C()' for categorical treatment.
        # 'birth_year' is now in control_vars and thus not wrapped in C().
        formula = f"{outcome} ~ {elig} + {' + '.join(control_vars)} + {' + '.join([f'C({fe})' for fe in fixed_effect_vars])}"

        try:
            # Run a Weighted Least Squares (WLS) regression
            # Set weights equal to data['ASECWT']. Cluster standard errors by STATEFIP.
            model = smf.wls(formula=formula, data=data, weights=data['ASECWT']).fit(cov_type='cluster', cov_kwds={'groups': data['STATEFIP']})

            # Extract the coefficient, standard error, p-value for the elig variable
            coeff = model.params[elig]
            std_err = model.bse[elig]
            p_value = model.pvalues[elig]
            r_squared = model.rsquared_adj # Using adjusted R-squared for models with many predictors
            n_obs = model.nobs

            # Append a dictionary containing these results to results_list.
            results_list.append({
                'Outcome': outcome,
                'Eligibility': elig,
                'Coefficient': coeff,
                'Std Error': std_err,
                'P-value': p_value,
                'R-squared': r_squared,
                'N': int(n_obs)
            })

        except Exception as e:
            # Inside the except block, print an error message indicating which outcome/eligibility combination failed.
            print(f"Error running regression for Outcome: {outcome}, Eligibility: {elig} - {e}")

# 9. Convert results_list into a pandas DataFrame called results_df.
results_df = pd.DataFrame(results_list)

print("Regression results DataFrame re-created successfully with updated variables.")
print(results_df.head())
```

**Reasoning**:
Now that the regressions have been re-run with the updated variable lists and the results are stored in `results_df`, the next step is to format these results, create a pivot table for better readability, and then display both the detailed and pivoted results.



```python
import pandas as pd

# 10. Define a function, add_stars(p_value)
def add_stars(p_value):
    if p_value < 0.01:
        return '***'
    elif p_value < 0.05:
        return '**'
    elif p_value < 0.1:
        return '*'
    else:
        return ''

# 11. Create a new column in results_df called 'Formatted Result'
results_df['Formatted Result'] = results_df.apply(
    lambda row: f"{row['Coefficient']:.3f}{add_stars(row['P-value'])} ({row['Std Error']:.3f})",
    axis=1
)

# 12. Create a pivot table called pivot
pivot = results_df.pivot_table(
    index='Outcome',
    columns='Eligibility',
    values='Formatted Result',
    aggfunc='first' # Specify aggfunc to avoid aggregation error on string values
)

# Ensure the columns are ordered as specified
pivot = pivot[['cumulative_eligibility_at_age', 'elig_0_5', 'elig_6_11', 'elig_12_18']]

# 13. Display pivot and results_df
print("\n--- Regression Results Pivot Table ---")
print(pivot)

print("\n--- Detailed Regression Results DataFrame ---")
print(results_df)
```

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import warnings
warnings.filterwarnings("ignore")

# Force a CLEAN copy with NO category dtypes
cols_needed = ['mom_employed', 'dad_employed', 'mom_weeks', 'mom_hours',
               'cumulative_eligibility_at_age', 'elig_0_5', 'elig_6_11', 'elig_12_18',
               'Childhood_Avg_Unemployment', 'log_childhood_income', 'birth_year',
               'STATEFIP', 'age_child', 'SEX', 'RACE', 'HISPAN', 'ASECWT']
cols_available = [c for c in cols_needed if c in df_merged.columns]

df_parent = pd.DataFrame()
for col in cols_available:
    df_parent[col] = pd.to_numeric(df_merged[col], errors='coerce')

print(f"df_parent: {df_parent.shape}")
print(f"Dtypes:\n{df_parent.dtypes}")

parent_outcomes = [v for v in ['mom_employed', 'dad_employed', 'mom_weeks', 'mom_hours']
                   if v in df_parent.columns and df_parent[v].notna().sum() > 1000]
eligibility_vars = ['cumulative_eligibility_at_age', 'elig_0_5', 'elig_6_11', 'elig_12_18']

print(f"\nOutcomes: {parent_outcomes}")

results_list = []

for outcome in parent_outcomes:
    for elig in eligibility_vars:
        formula = (f"{outcome} ~ {elig} + Childhood_Avg_Unemployment + log_childhood_income"
                   f" + birth_year + C(STATEFIP) + C(age_child) + C(SEX) + C(RACE) + C(HISPAN)")

        data = df_parent[[outcome, elig, 'Childhood_Avg_Unemployment', 'log_childhood_income',
                          'birth_year', 'STATEFIP', 'age_child', 'SEX', 'RACE', 'HISPAN',
                          'ASECWT']].dropna().copy()
        data = data[data['ASECWT'] > 0].reset_index(drop=True)

        if len(data) < 100:
            continue

        try:
            model = smf.wls(formula=formula, data=data, weights=data['ASECWT'])
            res = model.fit(cov_type='cluster', cov_kwds={'groups': data['STATEFIP']})

            results_list.append({
                'Outcome': outcome, 'Eligibility': elig,
                'Coefficient': res.params[elig], 'Std Error': res.bse[elig],
                'P-value': res.pvalues[elig], 'N': int(res.nobs)
            })
            print(f"  OK: {outcome} ~ {elig}: coef={res.params[elig]:.4f}, p={res.pvalues[elig]:.4f}")
        except Exception as e:
            print(f"  ERROR: {outcome} ~ {elig}: {e}")

# Format
results_df_parent = pd.DataFrame(results_list)

def add_stars(p):
    if pd.isna(p): return ''
    if p < 0.01: return '***'
    if p < 0.05: return '**'
    if p < 0.1: return '*'
    return ''

results_df_parent['Formatted'] = results_df_parent.apply(
    lambda r: f"{r['Coefficient']:.4f}{add_stars(r['P-value'])} ({r['Std Error']:.4f})", axis=1)

pivot = results_df_parent.pivot_table(
    index='Outcome', columns='Eligibility', values='Formatted', aggfunc='first')
col_order = ['cumulative_eligibility_at_age', 'elig_0_5', 'elig_6_11', 'elig_12_18']
pivot = pivot[[c for c in col_order if c in pivot.columns]]

print("\n" + "="*80)
print("PARENTAL LABOR SUPPLY RESULTS")
print("="*80)
print(pivot)
```
